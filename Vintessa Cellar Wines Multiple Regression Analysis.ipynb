# Breusch-Pagan Test for Homoscedasticity
This test is mainily for heteroscedasticity but the result can also tell if it has homoscedasticity by the level of the p-value. High p-value means **homoscedasticity** and a low p-value means **heteroscedasticity**

```python
bre_test=het_breuschpagan(reg_model_refined.resid, reg_model_refined.model.exog)
bre_labels=['Lagrange multiplier statistic','p-value','f-value','fp-value']

print("Breusch-Pagan Test Results:")
for label, result in zip(bre_labels, bre_test):
  print(f"{label}:{result}")
```

```python
if bre_test[1]>0.05:
  print("Homoscedasticity holds")
else:
  print("Homoscedasticity does not hold")
```
